```python
# データの読み込み
file_path = "For_Transaction_data_Comparison.csv"
df = pd.read_csv(file_path)
df.head()
```

```python
# Specify the columns for which you want to calculate statistics
columns_of_interest = ['negative_count_1','negative_count_2','negative_count_3','negative_count_4']

# Calculate basic statistics for the specified columns
statistics = df[columns_of_interest].describe()

# Display the statistics
# 統計量の名前を指定
#statistics.loc['Statistic'] = ['count', 'mean', 'std', 'min', '25%', '50%', '75%', 'max']

# 四分位偏差を計算して統計量に追加
q1 = df[columns_of_interest].quantile(0.25)
q3 = df[columns_of_interest].quantile(0.75)
iqr = q3 - q1

# 統計量の行に四分位偏差を追加
statistics.loc['IQR'] = iqr
#statistics.loc['q3_dev'] = q3

print(statistics)
# 結果をCSVファイルに書き込む
statistics.to_csv('BasicStatist/negative_count_BA.csv', index=False,encoding='shift-jis')  # 出力先のファイル名に適宜変更
```
