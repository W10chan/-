```python
import pandas as pd
from scipy.stats import spearmanr

# CSVファイルを読み込む
df = pd.read_csv('4th_trading_trend_2023.csv')

# 列の選択
selected_columns = ['Med_purchase_num','Max_purchase_num', 'Total number of shares purchased', 'Med_disposal_num',
                     'Max_disposal_num', 'Total number of shares sold', 'Kd', 'Ki']

# データフレームから必要な列を抽出
selected_df = df[selected_columns]

# 正のKdに対するスピアマンの相関係数
correlation_positive_kd = selected_df[selected_df['Kd'] > 0].corr(method='spearman')

# 負のKdに対するスピアマンの相関係数
correlation_negative_kd = selected_df[selected_df['Kd'] < 0].corr(method='spearman')

# 正のKiに対するスピアマンの相関係数
correlation_positive_ki = selected_df[selected_df['Ki'] > 0].corr(method='spearman')

# 負のKiに対するスピアマンの相関係数
correlation_negative_ki = selected_df[selected_df['Ki'] < 0].corr(method='spearman')

# 結果を新しいCSVファイルに記録
result_df = pd.DataFrame({
    'Correlation_Positive_Kd': round(correlation_positive_kd.loc['Kd'], 2),
    'Correlation_Negative_Kd': round(correlation_negative_kd.loc['Kd'], 2),
    'Correlation_Positive_Ki': round(correlation_positive_ki.loc['Ki'], 2),
    'Correlation_Negative_Ki': round(correlation_negative_ki.loc['Ki'], 2)
})

result_df.to_csv('4th_cons_sign.csv', index=False)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import spearmanr

# CSVファイルを読み込む
df = pd.read_csv('1st_trading_trend_2023.csv')

# 列の選択
selected_columns = ['Med_purchase_num','Max_purchase_num', 'Total number of shares purchased', 'Med_disposal_num',
                     'Max_disposal_num', 'Total number of shares sold', 'Kd', 'Ki']

# データフレームから必要な列を抽出
selected_df = df[selected_columns]

# 正のKdに対するスピアマンの相関係数
correlation_positive_kd = selected_df[selected_df['Kd'] > 0].corr(method='spearman')

# 負のKdに対するスピアマンの相関係数
correlation_negative_kd = selected_df[selected_df['Kd'] < 0].corr(method='spearman')

# 正のKiに対するスピアマンの相関係数
correlation_positive_ki = selected_df[selected_df['Ki'] > 0].corr(method='spearman')

# 負のKiに対するスピアマンの相関係数
correlation_negative_ki = selected_df[selected_df['Ki'] < 0].corr(method='spearman')

# ヒートマップを描く
plt.figure(figsize=(10, 8))

# 正のKd
#plt.subplot(2, 2, 1)
sns.heatmap(correlation_positive_kd, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation - Positive Kd')
plt.tight_layout()
plt.show()

# ヒートマップを描く
plt.figure(figsize=(10, 8))
# 負のKd
#plt.subplot(2, 2, 2)
sns.heatmap(correlation_negative_kd, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation - Negative Kd')
plt.tight_layout()
plt.show()

# ヒートマップを描く
plt.figure(figsize=(10, 8))
# 正のKi
#plt.subplot(2, 2, 3)
sns.heatmap(correlation_positive_ki, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation - Positive Ki')
plt.tight_layout()
plt.show()

# ヒートマップを描く
plt.figure(figsize=(10, 8))
# 負のKi
#plt.subplot(2, 2, 4)
sns.heatmap(correlation_negative_ki, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation - Negative Ki')

plt.tight_layout()
plt.show()
```
